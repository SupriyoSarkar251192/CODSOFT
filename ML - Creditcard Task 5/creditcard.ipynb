## Importing necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## Loading the Dataset

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

## Checking for Duplicate and Null values

```python
df.duplicated().sum()
```

```python
df.isna().sum().values
```

Insight: There is no null value in this dataset, but it contains 1081 duplicate values.

### Removing duplicated values

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

Insight: Class is a kind of categorical feature with numeric values 0 and 1 which determines whethere the transaction is fraud or not.

Amount feature is a widely distributed feature as it shows its mean value much smaller than its standard daviation. So there is a high chance of having a large number of outliers.

```python
df['Class'].value_counts()
```

Insight: It seems like it is typically imbalanced.

### Downsampling the dataset for making it balanced

```python
df_nonfraud = df[df['Class'] == 0]
df_fraud = df[df['Class'] == 1]
```

```python
from sklearn.utils import resample

df_nonfraud_downsampled = resample(
    df_nonfraud, 
    replace=False,
    n_samples= len(df_fraud),
    random_state= True
)
```

```python
df_resampled = pd.concat([df_nonfraud_downsampled, df_fraud])
df_resampled.shape
```

### Checking for outliers

```python
features = df_resampled.columns.drop('Class')

len(features)
```

```python
fig, axes = plt.subplots(15, 2, figsize=(8, 45))
axes = axes.flatten()

for i in range(30):
    sns.boxplot(df_resampled, x=features[i], ax=axes[i])

plt.tight_layout()
plt.show()

    
```

Insight: There are negative outliers in all features except amount. So its better to use capping technique to handle outliers.

## Capping outliers

```python
def capping(data: pd.DataFrame, feature):
    data1 = data.copy()
    for i in features:
        q1 = data1[i].quantile(0.25)
        q3 = data1[i].quantile(0.75)
        iqr = q3 - q1
        lr = q1 - iqr * 1.5
        ur = q3 + iqr * 1.5
        data1.loc[data1[i] < lr,i] = lr
        data1.loc[data1[i] > ur,i] = ur
    return data1

features.drop('Time')
df_capped = capping(df_resampled, features)
```

```python
fig, axes = plt.subplots(15, 2, figsize=(8, 45))
axes = axes.flatten()

for i in range(30):
    sns.boxplot(df_capped, x=features[i], ax=axes[i])

plt.tight_layout()
plt.show()
```

## Splitting of data

```python
x = df_capped.drop(columns=['Class'])
y = df_capped['Class']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Model training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV

clf = RandomForestClassifier()
params = {
    'max_depth': [1, 2, 3, 5, 10, None],
    'n_estimators': [50, 100, 200, 300],
    'criterion': ['gini', 'entropy']
}

model = GridSearchCV(clf, param_grid=params, cv=5, verbose=3, scoring='accuracy')
model.fit(x_train, y_train)
```

```python
model.best_score_, model.best_params_
```

## Prediction

```python
y_pred = model.predict(x_test)

print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
print(f"Precision Score: {precision_score(y_test, y_pred)}")
print(f"Recall Score: {recall_score(y_test, y_pred)}")
print(f"f1 Score: {f1_score(y_test, y_pred)}")
print(classification_report(y_test, y_pred))
```
